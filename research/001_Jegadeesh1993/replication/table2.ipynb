```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# ============================================================
# Table 2 Beta Notebook
# 6/3 momentum, no market cap for now
# ============================================================

# ----------------------------
# 1) Prepare monthly data
# ----------------------------
daily_close = prices["Close"].copy()
daily_close = daily_close.sort_index()
daily_close = daily_close.dropna(axis=1, how="all")

monthly_close = daily_close.resample("ME").last()
monthly_close = monthly_close.dropna(axis=1, how="all")

# monthly returns
monthly_ret = monthly_close.pct_change(fill_method=None)

print("monthly_close shape:", monthly_close.shape)
print("monthly_ret shape:", monthly_ret.shape)
print("monthly_close index head:", monthly_close.index[:5])
print("monthly_ret index head:", monthly_ret.index[:5])

# equal-weighted market proxy
market_ret = monthly_ret.mean(axis=1, skipna=True).rename("market_ret")


# ----------------------------
# 2) Helpers
# ----------------------------
def compute_formation_signal(monthly_close: pd.DataFrame, J: int = 6) -> pd.DataFrame:
    """
    Past J-month cumulative return used for ranking.
    signal[t] = P_t / P_{t-J} - 1
    """
    return monthly_close.pct_change(periods=J, fill_method=None)


def assign_deciles(signal_row: pd.Series, n_deciles: int = 10) -> pd.Series:
    """
    1 = losers decile
    10 = winners decile
    """
    x = signal_row.dropna()
    if len(x) < n_deciles:
        return pd.Series(dtype="int64")

    # Handle ties robustly
    ranks = x.rank(method="first")
    labels = pd.qcut(ranks, n_deciles, labels=False) + 1
    return pd.Series(labels.astype(int), index=x.index)


def build_overlapping_decile_returns(
    monthly_close: pd.DataFrame,
    monthly_ret: pd.DataFrame,
    J: int = 6,
    K: int = 3,
    n_deciles: int = 10,
) -> pd.DataFrame:
    """
    Build overlapping decile portfolio returns.

    For each formation month:
    - rank stocks by past J-month return
    - assign deciles 1..10
    - hold each decile portfolio for K months
    - at each calendar month, average across active sub-portfolios

    Returns a DataFrame with columns:
    P1 ... P10   : decile returns
    N1 ... N10   : average number of names in each decile (optional diagnostic)
    """
    signal = compute_formation_signal(monthly_close, J=J)
    dates = monthly_close.index

    subportfolios = []
    for formation_pos, dt in enumerate(dates):
        if dt not in signal.index:
            continue

        row = signal.loc[dt]
        deciles = assign_deciles(row, n_deciles=n_deciles)
        if deciles.empty:
            continue

        decile_names = {
            d: deciles[deciles == d].index.tolist()
            for d in range(1, n_deciles + 1)
        }

        start_pos = formation_pos + 1
        end_pos = min(formation_pos + K, len(dates) - 1)  # inclusive

        if start_pos >= len(dates):
            continue

        subportfolios.append(
            {
                "formation_date": dt,
                "start_pos": start_pos,
                "end_pos": end_pos,
                "decile_names": decile_names,
            }
        )

    out_rows = []
    for t_pos, dt in enumerate(dates):
        active = [sp for sp in subportfolios if sp["start_pos"] <= t_pos <= sp["end_pos"]]
        if len(active) == 0:
            continue

        ret_row = monthly_ret.loc[dt]

        record = {
            "date": dt,
            "n_active_subportfolios": len(active),
        }

        for d in range(1, n_deciles + 1):
            decile_month_returns = []
            decile_sizes = []

            for sp in active:
                names = sp["decile_names"][d]
                if len(names) == 0:
                    continue

                r = ret_row[names].dropna().mean()
                if pd.notna(r):
                    decile_month_returns.append(r)
                    decile_sizes.append(len(names))

            record[f"P{d}"] = float(np.mean(decile_month_returns)) if len(decile_month_returns) > 0 else np.nan
            record[f"N{d}"] = float(np.mean(decile_sizes)) if len(decile_sizes) > 0 else np.nan

        out_rows.append(record)

    decile_returns = pd.DataFrame(out_rows).set_index("date").sort_index()
    return decile_returns


def estimate_beta_table(
    decile_returns: pd.DataFrame,
    market_ret: pd.Series,
    nw_lags: int = 5,
) -> pd.DataFrame:
    """
    Estimate post-ranking beta for each decile portfolio:
        P_d,t = alpha_d + beta_d * market_t + error_t

    Uses HAC / Newey-West standard errors.
    """
    df = decile_returns.join(market_ret, how="inner").dropna(subset=["market_ret"]).copy()

    rows = []
    for d in range(1, 11):
        y = df[f"P{d}"].dropna()
        x = df.loc[y.index, "market_ret"]
        X = sm.add_constant(x)

        model = sm.OLS(y, X).fit(cov_type="HAC", cov_kwds={"maxlags": nw_lags})

        rows.append(
            {
                "decile": f"P{d}",
                "alpha": model.params["const"],
                "beta": model.params["market_ret"],
                "t_alpha": model.tvalues["const"],
                "t_beta": model.tvalues["market_ret"],
                "p_beta": model.pvalues["market_ret"],
                "n_obs": int(model.nobs),
                "mean_return": y.mean(),
                "std_return": y.std(ddof=1),
            }
        )

    beta_table = pd.DataFrame(rows).set_index("decile")
    return beta_table


# ----------------------------
# 3) Run 6/3 strategy
# ----------------------------
J = 6
K = 3

decile_returns_63 = build_overlapping_decile_returns(
    monthly_close=monthly_close,
    monthly_ret=monthly_ret,
    J=J,
    K=K,
    n_deciles=10,
)

print("\nDecile returns head:")
display(decile_returns_63.head())

beta_table_63 = estimate_beta_table(
    decile_returns=decile_returns_63,
    market_ret=market_ret,
    nw_lags=5,
)

print("\nBeta table (6/3):")
display(beta_table_63.round(4))

# ----------------------------
# 4) Convenience columns
# ----------------------------
beta_table_63["rank"] = np.arange(1, 11)
beta_table_63["decile_num"] = beta_table_63["rank"]

# Winners minus losers beta proxy (difference in average returns, not true hedge beta)
P1 = decile_returns_63["P1"]
P10 = decile_returns_63["P10"]
wml = (P10 - P1).dropna()

wml_model = sm.OLS(
    wml,
    sm.add_constant(market_ret.loc[wml.index]),
).fit(cov_type="HAC", cov_kwds={"maxlags": 5})

wml_summary = pd.Series(
    {
        "alpha": wml_model.params["const"],
        "beta": wml_model.params["market_ret"],
        "t_alpha": wml_model.tvalues["const"],
        "t_beta": wml_model.tvalues["market_ret"],
        "p_beta": wml_model.pvalues["market_ret"],
        "n_obs": int(wml_model.nobs),
        "mean_return": wml.mean(),
        "std_return": wml.std(ddof=1),
    },
    name="P10-P1",
)

print("\nWinners minus losers (P10-P1):")
display(wml_summary.to_frame().T.round(4))

# ----------------------------
# 5) Clean final table
# ----------------------------
final_table = beta_table_63[["beta", "t_beta", "mean_return", "n_obs"]].copy()
final_table.loc["P10-P1"] = [
    wml_summary["beta"],
    wml_summary["t_beta"],
    wml_summary["mean_return"],
    wml_summary["n_obs"],
]

print("\nFinal table:")
display(final_table.round(4))

# ----------------------------
# 6) Plot beta by decile
# ----------------------------
plt.figure(figsize=(10, 5))
plt.plot(
    range(1, 11),
    beta_table_63["beta"].values,
    marker="o",
)
plt.xticks(range(1, 11), [f"P{i}" for i in range(1, 11)])
plt.xlabel("Decile")
plt.ylabel("Post-ranking beta")
plt.title("6/3 Momentum: Post-ranking Beta by Decile")
plt.grid(True, alpha=0.3)
plt.show()

# ----------------------------
# 7) Optional diagnostics
# ----------------------------
print("\nAverage decile returns:")
display(decile_returns_63[[f"P{i}" for i in range(1, 11)]].mean().to_frame("mean_return").round(4))

print("\nAverage number of names per decile:")
display(decile_returns_63[[f"N{i}" for i in range(1, 11)]].mean().to_frame("avg_names").round(2))
```
